```python
import math
# Define masses in atomic mass units (amu)
m1 = 1.00727647   # Proton rest mass
m2 = 12           # Carbon-12 rest mass
amu = 931.5
exc = 4.438

m1_rel = m1 * amu
m2_rel = m2 * amu
m3_rel = m2_rel + exc

qval = -exc

E_b = 14.77
E_cm = E_b * (m2/(m1+m2))
y = m1_rel + m2_rel
beta = math.sqrt(E_b*(E_b+2*m1_rel))/(y + E_b)
ecmi =  math.sqrt(y**2 + 2*E_b*m2_rel)
ecmf =  ecmi + qval - y + m1_rel + m3_rel
e3cm = (ecmf**2 + (m1_rel + m3_rel)*(m1_rel - m3_rel))/(2*ecmf)
beta3c = math.sqrt(1-1/(e3cm/m1_rel)**2)
y_new = ((e3cm/m1_rel)**2)*(1-beta**2)
c = 1 - y_new

def lab_energy(theta):
    cos_theta = np.cos(theta)
    b = -beta*cos_theta
    a = y_new + b**2
    d = b**2 - a*c
    e = (-b + d**0.5)/a
    return m1_rel * ((1/((1-e**2)**0.5)-1))


# Array of scattering angles
theta_arr = np.linspace((np.pi)/4, (np.pi)/2, 100)

# Array of lab energies
E_lab_arr = lab_energy(theta_arr)

# Plot lab energy as a function of lab angle
plt.plot(theta_arr * 180/np.pi, E_lab_arr)
plt.xlabel('Scattering Angle (degrees)')
plt.ylabel('Lab Energy (MeV)')
plt.title('Lab Energy vs. Lab Angle for 14.77 MeV Protons on Carbon-12')
plt.show()
```

```python
import math
# Define masses in atomic mass units (amu)
m1 = 1.00727647   # Proton rest mass
m2 = 12           # Carbon-12 rest mass
m2 = 95.90468
amu = 931.5
exc = 2.1

m1_rel = m1 * amu
m2_rel = m2 * amu
m3_rel = m2_rel + exc

qval = -exc

E_b = 14.83
E_cm = E_b * (m2/(m1+m2))
y = m1_rel + m2_rel
beta = math.sqrt(E_b*(E_b+2*m1_rel))/(y + E_b)
ecmi =  math.sqrt(y**2 + 2*E_b*m2_rel)
ecmf =  ecmi + qval - y + m1_rel + m3_rel
e3cm = (ecmf**2 + (m1_rel + m3_rel)*(m1_rel - m3_rel))/(2*ecmf)
beta3c = math.sqrt(1-1/(e3cm/m1_rel)**2)
y_new = ((e3cm/m1_rel)**2)*(1-beta**2)
c = 1 - y_new

def lab_energy(theta):
    cos_theta = np.cos(theta)
    b = -beta*cos_theta
    a = y_new + b**2
    d = b**2 - a*c
    e = (-b + d**0.5)/a
    return m1_rel * ((1/((1-e**2)**0.5)-1))


# Array of scattering angles
theta_arr = np.linspace(0, np.pi, 100)

# Array of lab energies
E_lab_arr = lab_energy(theta_arr)

fig, ax = plt.subplots(figsize=(8, 6))

ax.hist2d(df_barrel['ThetaLab'],df_barrel['TotalEnergy']
        ,bins=[180,1000], norm=mpl.colors.LogNorm(), cmap='rainbow', cmin=1)

ax.plot(theta_arr * 180/np.pi, E_lab_arr*1000, label = '4.44 MeV state', linewidth=1, color = 'b')


ax.set_xlabel('Scattering Angle (degrees)')
ax.set_ylabel('Lab Energy (MeV)')
#ax.set_ylim(8000,10000)
ax.legend()
plt.title('Lab Energy vs. Lab Angle')
plt.show()
```
